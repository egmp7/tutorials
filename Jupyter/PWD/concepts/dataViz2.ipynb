```python
import pandas as pd

inputfile = "https://dr3vr6j2erh62.cloudfront.net/mediastore/dsm020/datasets/IRIS.csv"
df = pd.read_csv(inputfile)
df.head()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
species = sorted(list(set (df['species'])))
counts = [df.loc[df['species'] == name].size for name in species]
explode = (0,0.1,0.2)

plt.figure()
plt.pie(counts,
       labels= species,
       explode = explode,
       autopct = '%1.2f')
plt.show()
```

```python
plt.figure()
plt.hist(df["sepal_length"],bins = 30)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
for specie in species:
    data = df.loc[df['species']==specie]
    sns.distplot(data['sepal_length'],
               hist = True,
               kde = True,
               bins = 30,
               label = specie,
               kde_kws = {'bw': .1}
               )



plt.show()
```

```python
plt.figure()
plt.scatter(df['sepal_length'],
           df['sepal_width'],
           s = 10)
plt.show()
```

```python
species_color = {
    "setosa": "green",
    "virginica": "orange",
    "versicolor": "purple"
}

plt.figure()
plt.scatter(df['sepal_length'],
            df['sepal_width'],
            s = 10,
            c = [species_color[v] for v in df['species']])
plt.title("Sepal length vs Sepal width")
plt.xlabel("Sepal Width")
plt.ylabel("Sepal Length")
plt.show()
```

```python
plt.figure()

for specie in species:
    data = df.loc[df['species'] == specie]
    plt.scatter(data['sepal_length'],
               data['sepal_width'],
               label = specie,
               s = 10,
               c = species_color[specie]
               )
plt.legend()
plt.show()
```
