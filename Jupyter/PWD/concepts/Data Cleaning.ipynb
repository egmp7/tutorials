```python
m = np.mean(df2["plasma_glucose_concentration"])
sd = np.std(df2["plasma_glucose_concentration"])
```

```python
df4 = df2.copy()
for (i,val) in enumerate (df4["plasma_glucose_concentration"]):
    if(val > (m +(sd * 2.5))) or (val < (m -(sd * 2.5))):
        df4.at[i,"plasma_glucose_concentration"] = m
        print("replaced an outlier: ",i, val)
```

<h3>2. Data Cleaning</h3>

```python
df5 = df2.copy()
m = np.mean(df2["plasma_glucose_concentration"])
sd = np.std(df2["plasma_glucose_concentration"])

df5.loc[df5["plasma_glucose_concentration"] > ((m +(sd * 2.5))), "plasma_glucose_concentration"] = m
df5.loc[df5["plasma_glucose_concentration"] < ((m -(sd * 2.5))), "plasma_glucose_concentration"] = m
df5.describe()
```

<h3>Masks</h3>

```python
#1,2,3 = false, true , false = false , 2 , false

df6 = df2.copy()
m = np.mean(df2["plasma_glucose_concentration"])
sd = np.std(df2["plasma_glucose_concentration"])
threshold = sd*2.5

mask = (df6["plasma_glucose_concentration"] > (m + threshold)) | \
        (df6["plasma_glucose_concentration"] < (m - threshold))
df6.loc[mask,"plasma_glucose_concentration"]= m
df6.describe()
```
