```python
df2 = df.copy()
df2 = df2.loc[df["Tm"] == "CHI"]
df2
```

```python
dummies = pd.get_dummies(df2['Pos'])
dummies
```

```python
df3 = pd.concat([df2,dummies], axis = 1 , sort = True)
df3.drop('Pos', axis = 1, inplace = True)
df3
```

```python
import numpy as np

df4 = df3.copy()
df4['log(MP)'] = np.log(df4['MP'])
df4
```

```python
def rescale_zero2one(x, lo, hi):
    return(x-lo) / (hi - lo)

s = pd.Series([-1,0.1,1.,3.])
s = s.apply(rescale_zero2one, args=(s.min(),s.max()))
s
```

```python
lo = df4['VORP'].min()
hi = df4['VORP'].max()

df4['scaled(VORP)'] = df4['VORP'].apply(rescale_zero2one, args = (lo,hi))
df4
```
